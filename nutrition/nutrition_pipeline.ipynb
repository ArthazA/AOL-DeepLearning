```python
import json

with open("nutrition_per_100g.json") as f:
    nutrition_100g = json.load(f)

with open("avg_weights.json") as f:
    avg_weights = json.load(f)

with open("../nutrition/class_id_map.json") as f:
    CLASS_ID_MAP = json.load(f)
```

```python
def compute_nutrition(counts):
    summary = {}
    totals = {"calories": 0, "carbs": 0, "protein": 0, "fat": 0, "fiber": 0}

    for item, count in counts.items():
        weight = avg_weights[item] * count
        factor = weight / 100
        per100 = nutrition_100g[item]

        summary[item] = {
            "count": count,
            "weight_g": weight,
            "calories": per100["calories"] * factor,
            "carbs": per100["carbs"] * factor,
            "protein": per100["protein"] * factor,
            "fat": per100["fat"] * factor,
            "fiber": per100["fiber"] * factor
        }

        for k in totals:
            totals[k] += summary[item][k]

    return summary, totals
```

```python
summary, totals = compute_nutrition({'apple': 2, 'banana': 1})
summary, totals
```
