```python
from ultralytics import YOLO
from pathlib import Path
```

```python
# import torch
# torch.cuda.empty_cache()
```

```python
def main():
    PROJECT_ROOT = Path.cwd().parent

    # Change path here
    DATA_YAML = PROJECT_ROOT / "data" / "FNVdataset_unmodified" / "data.yaml" # change "FNVdataset_unmodified" to your file name
    RUNS_DIR = PROJECT_ROOT / "runs"

    model = YOLO("yolov8s.pt") # pretrained in COCO weights

    model.train(
        data=str(DATA_YAML),
        epochs=100,
        imgsz=640,
        batch=-1,
        optimizer="AdamW",
        lr0=1e-3,
        cos_lr=True,
        weight_decay=5e-4,

        label_smoothing=0.05,
        patience=8,

        hsv_h=0.015,
        hsv_s=0.7,
        hsv_v=0.4,
        degrees=5.0,
        translate=0.1,
        scale=0.5,
        shear=0.0,
        flipud=0.0,
        fliplr=0.5,
        mosaic=0.5,
        mixup=0.0,

        workers=8,
        device=0,
        amp=True,

        # freeze=10,

        project=str(RUNS_DIR),
        name="yolov8s_fnv_correct_p3", # change "yolov8s_fnv_correct_p3" to your file name
        exist_ok=False # change to true if directory exists
    )

    print("\nTraining completed successfully.")
    print("Best model saved in:")
    print(RUNS_DIR / "yolov8s_fnv_correct_p3" / "weights" / "best.pt") # change "yolov8s_fnv_correct_p3" to your file name


if __name__ == "__main__":
    main()
```
